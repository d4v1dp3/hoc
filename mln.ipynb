## Introducción
Es un paradigma dentro del aprendizaje maquina donde los datos de entrada y la salida deseada son etiquetados para el entrenamiento. Estos datos de entrenamiento suelen ser etiquetados por un científico de datos en la fase de preparación, antes de ser utilizados para entrenar y probar el modelo. Una vez que el modelo ha aprendido la relación entre los datos de entrada y de salida, puede utilizarse para clasificar conjuntos de datos nuevos y desconocidos y predecir resultados.

### Clasificación
La clasificación es una tarea que consiste en identificar los datos de entrada como parte de un grupo aprendido y asignar una etiqueta de clase.

<img style="max-width:400px; display:block;margin-left: auto; margin-right: auto;" src="https://drive.google.com/uc?export=view&id=1Sov36Jab9VtK3Pgq-KbBgjo_ZJpXecLf" alt="">

Algunas aplicaciones que se le pueden dar son:
* Identificación y clasificación de objetos en una imagen.
* Reconocimiento facial y de voz.
* Detección de spam en algún filtro de correo.

Existen diferentes clasificaciones de este tipo de problemas, donde generalmente depende de la cantidad de clases que hay que aplicar a los datos en un entorno real.

Una clasificación simplificada podemos encontrar:
* Binaria: La clasificación binaria se produce cuando un modelo sólo puede aplicar dos etiquetas de clase. El uso más popular para este tipo de clasificación podría ser en detectar correos basura. Algunos algoritmos que implementan lo implementan:
    * Regresión logistica
    * Arboles de decisión
    * Naïve Bayes

<img style="max-width:400px; display:block;margin-left: auto; margin-right: auto;" src="https://drive.google.com/uc?export=view&id=1xkMG0KG2a_O6l4Lzd9XmQZy8UppH1krl" alt="">

* Múltiples clases: La clasificación de clases múltiples es cuando los modelos hacen referencia a mas de las dos etiquetas de clase que se encuentran en la clasificación binaria. En su lugar, puede haber una enorme variedad de posibles etiquetas de clase que pueden aplicarse al objeto o a los datos. Un ejemplo sería el software de reconocimiento facial, en el que un modelo puede analizar una imagen comparándola con una amplia gama de posibles etiquetas de clase para identificar al individuo.
    * Random Forest
    * K-NN (Nearest Neighbors)
    * Naïve Bayes 
* Múltiples etiquetas: La clasificación de etiquetas múltiples se produce cuando el modelo de aprendizaje automático puede asignar a un objeto o punto de datos más de una etiqueta de clase. En este caso, el modelo suele tener varias salidas. Un ejemplo de esto podría ser la clasificación de imágenes, donde pueden contener múltiples objetos.
    * Gradient Boosting
    * Random Forests
    * Usar diferentes algoritmos de clasificación para cada etiqueta de clase.


### Regresión
La regresión consiste en predecir resultados a partir de datos en continuo cambio. Es una técnica para investigar la relación entre variables independientes o características y una variable dependiente o resultado.

<img style="max-width:400px; display:block;margin-left: auto; margin-right: auto;" src="https://drive.google.com/uc?export=view&id=1hhJWXhL1p5og8XGPJHFxSvGnO7vbvky4" alt="">

Algunas aplicaciones que se le pueden dar:
* Predecir resultados bursátiles o comerciales
* Predicción del mercado inmobiliario
* Anuncios

Una clasificación simplificada podemos encontrar:
* Regresión lineal simple: Es un proceso popular de enfoque de regresión y se utiliza para predecir la salida objetivo a partir de una variable de entrada. Una vez que se ha entrenado un modelo sobre la relación entre la entrada y la salida objetivo, puede utilizarse para hacer predicciones sobre nuevos datos.
* Regresión basada en arboles de decisión: Los modelos de árbol de decisión adoptan la estructura de un árbol en el que el modelo se ramifica de forma incremental. Puede utilizarse tanto para regresión como clasificación. El conjunto de datos se divide en subconjuntos incrementales y puede utilizarse para comprender la correlación entre variables independientes.

# Ventajas
* El aprendizaje supervisado permite recopilar datos y produce resultados a partir de experiencias anteriores.
* Puede realizar tareas de clasificación y regresión
* Permite estimar o asignar el resultado a una nueva muestra.
* Se tiene un control total sobre la elección del número de clases que queremos en los datos de entrenamiento.

# Desventajas
* Clasificar un conjunto de datos inmenso puede ser desafiante.
* Requiere un conjunto de datos etiquetado previamente.
* No puede agrupar datos por sí mismo.
* Datos inesperados disminuyen la eficacia.
* Si la entrada no coincide con alguna clase a partir del conjunto de entrenamiento,  el resultado podría ser una etiqueta de clase errónea tras la clasificación.
* Se puede sobreajustar (overfit) el algoritmo con facilidad.

El aprendizaje supervisado utiliza datos etiquetados para entrenar algoritmos de regresión o clasificación de datos que tienen aplicación en todos los dominios e industrias. Ademas que requiere científicos de datos experimentados para construir, escalar y actualizar los modelos.
